```python
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv("data/accepted_2007_to_2018Q4.csv")
```

```python
df.shape
```

```python
# The Output value is the load_status
df['loan_status'].value_counts()
```

```python
df = df[(df['loan_status']=='Fully Paid') | (df['loan_status']=='Charged Off')]
df['loan_status'].value_counts()
```

```python
df.shape
```

```python
# Need to clean up and check this huge data
# Checking for percentage of null values in columns
((df.isnull().sum()/len(df))*100).sort_values(ascending=False)
```

```python
# Plotting to check null values

plt.figure(figsize=(18,4),dpi=300) # make image larger and clearer
# y contains the % of missing values od each column and x is the index of the series in above cell
sns.barplot(y=((df.isnull().sum()/len(df))*100).sort_values(), x=((df.isnull().sum()/len(df))*100).sort_values().index, palette='viridis') 
plt.xticks(rotation=90);
```

```python
# The data set is huge, it'll be hardware inefficient to take 151 features into account. 
# So, we need to reduce the number of features drastically
# So dropping columns with more than 30% of missing datas. We could have filled with missing values, 
# but again, we need to reduce numbers of features 
keep = df.columns[((df.isnull().sum()/len(df))*100 < 30)].to_list()
keep
```

```python
len(keep)
# Reduced to 93 Features
```

```python
df = df[keep] # Dropping featuressss with greater than 30% null values
df.shape
```

```python
# Count Categorical and numerical features
df.dtypes.value_counts()
```

```python
# Check for categorical values
data = df.copy()
categorical = [feature for feature in data.columns if data[feature].dtype == "O"]
categorical
```

```python
# Check unique values of each categorical features
for value in categorical:
    print(value)
    print(data[value].nunique())
```

```python
# There are multiple columns that have too many unique values, we need to drop them. Dropping any column that have more than
# 40 unique values
for value in categorical:
    if data[value].nunique()>40:
        data.drop([value],axis=1,inplace=True)
```

```python
data.dtypes.value_counts()
```

```python
numerical = [feature for feature in data.columns if data[feature].dtype != "O"]
numerical
```

```python
# Copying the df to create another df with categorical feature converted to numeric
df_temp = data.copy()
df_temp['loan_status'] = pd.get_dummies(df_temp['loan_status'], drop_first=True)# Now Let's see which data it corelates into the best
```

```python
plt.figure(figsize=(10,10),dpi=120)
df_temp.corr()['loan_status'].sort_values().drop('loan_status').plot(kind='bar', cmap='viridis') # correlation with loan_status for continuous features with loan_status feature dropped
plt.xticks(rotation=90);
```
